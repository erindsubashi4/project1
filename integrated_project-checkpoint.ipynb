```python
!pip install joypy
import pandas as pd
import numpy as np
import joypy
games = pd.read_csv(r'C:\Users\User\Desktop\projects\integrated_project\games.csv')

games.columns = games.columns.str.lower()
games['year_of_release'] = games['year_of_release'].astype('float32')
games['jp_sales'] = games['jp_sales'].astype('float32')
games['eu_sales'] = games['eu_sales'].astype('float32')
games['na_sales'] = games['na_sales'].astype('float32')
games['other_sales'] = games['other_sales'].astype('float32')

games1 = games.dropna(axis=0, subset=['name']).reset_index()
games1['year_of_release'] = games['year_of_release'].fillna(games['year_of_release'].median())
rating_dict = {
    'Sports': 'E',
    'Platform': 'T',
    'Racing':'E',
    'Role-Playing':'M',
    'Puzzle':'AO',
    'Misc':'M',
    'Shooter':'E',
    'Simulation':'E10+',
    'Action':'M',
    'Fighting':'K-A',
    'Adventure':'EC',
    'Strategy':'RP'
}
games1['rating_new'] = games1.apply(lambda row: rating_dict[row['genre']] \
                                    if row['rating'] is np.nan else row['rating'], axis=1)
games1['critic_score'] = games1['critic_score'].fillna('undefined')
games1['user_score'] = games1['user_score'].fillna('undefined')
games1['total_sales'] = games1['na_sales'] + games1['eu_sales'] +games1['jp_sales'] +games1['other_sales']
print(games1)
```

```python
from scipy import stats as st


xbox = good_games.query('platform == "XOne"')
xbox = xbox[xbox['user_score'] != 'undefined']
xbox = xbox[xbox['user_score'] != 'tbd']
xbox['user_score'] = xbox['user_score'].astype('float')

pc = good_games.query('platform == "PC"')
pc = pc[pc['user_score'] != 'undefined']
pc = pc[pc['user_score'] != 'tbd']
pc['user_score'] = pc['user_score'].astype('float')
xbox_rating = xbox['user_score']
xbox_rating = xbox_rating.values
```

```python
pc_rating = pc['user_score']
pc_rating = pc_rating.values
```

```python
stat, p = results = st.ttest_ind(
        xbox_rating, 
        pc_rating, equal_var = False, nan_policy='omit')
print('p-value: ', results.pvalue)
alpha = 0.05
if p > alpha:
    print('There is no difference between XOne and PC, based on ESRB rating')
else:
    print('There is some difference between XOne and PC, based on ESRB rating')
```

Here I chosed H0: There is no difference between XOne and PC, based on ESRB rating.
H1: There is some difference between XOne and PC, based on ESRB rating. I chosed the student test. I also added equal_par = False as it is not significant difference between variances, and the nan_policy to omit.
